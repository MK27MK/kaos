```python
from kaos.data.enums import DataProvider
from kaos.data.instruments import FuturesContract
from kaos.data.loading import Catalog
import pandas as pd

# from kaos.visualization.visualization import hvplot_ohlc
from kaos.data.symbol import Symbol

catalog = Catalog()
symbols = [Symbol(f"E6-{m}-{y}") for m in "HMUZ" for y in range(2020, 2025)]
contracts: list[FuturesContract] = catalog.get_futures_contracts(
    symbols, DataProvider.FIRSTRATE, ["1day", "1min"], is_raw_data=True
)
print(*contracts, sep="\n")
```

```python
from kaos.data.enums import RolloverRule
from kaos.data.instruments import ContinuousFuturesContract


continuous = ContinuousFuturesContract.from_individuals(
    contracts, RolloverRule.OPEN_INTEREST
)
df = continuous.market_data["D"]
df
```
